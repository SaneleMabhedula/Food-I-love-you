```python
import matplotlib.pyplot as plt
import numpy as np

# Create price values
price = np.linspace(0, 10, 100)

# Demand function (downward slope)
demand = 20 - 2 * price

# Supply function (upward slope)
supply = 2 * price

# Plotting
plt.figure(figsize=(8,6))
plt.plot(price, demand, label="Demand", color="blue")
plt.plot(price, supply, label="Supply", color="red")

# Find equilibrium (where demand = supply)
equilibrium_price = 5
equilibrium_quantity = 10

# Mark equilibrium point
plt.scatter(equilibrium_price, equilibrium_quantity, color="black", zorder=5)
plt.text(equilibrium_price+0.3, equilibrium_quantity, "Equilibrium", fontsize=10)

# Labels
plt.title("Supply and Demand")
plt.xlabel("Price")
plt.ylabel("Quantity")
plt.legend()
plt.grid(True)
plt.show()
```
